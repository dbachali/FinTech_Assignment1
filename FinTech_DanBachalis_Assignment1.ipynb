```python
# Import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create x values for one period (2pi) in 0.1 rad increments
x = np.arange(-np.pi,np.pi,0.1)

# Create sine values
y = np.sin(x)

# Create cosine values
z = np.cos(x)

# Plot sine and cosine function
plt.plot(x,y,x,z)
plt.title('One Period of Sine and Cosine Functions')
plt.xlabel('x values (-\u03C0 to \u03C0)')
plt.ylabel('sin(x) and cos(x)')
plt.legend(['sin(x)', 'cos(x)'])    
plt.show()
```
