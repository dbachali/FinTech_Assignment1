```python
# Import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create x values for one period (2pi) in 0.1 rad increments
x = np.arange(-np.pi, np.pi, 0.1)

# Create sine values
s = np.sin(x)

# Create cosine values
c = np.cos(x)

# Create tangent values for one period (pi)
tx = np.arange(-np.pi/2, np.pi/2, 0.1)
t = np.tan(tx)

# Plot sine, cosine and tangent functions
plt.plot(x, s, x, c, tx, t)
plt.title('One Period of Sine, Cosine and Tangent Functions')
plt.xlabel('x values (-\u03C0 to \u03C0)')
plt.ylabel('sin(x), cos(x), tan(x)')
plt.legend(['sin(x)', 'cos(x)', 'tan(x)'])    
plt.ylim([-10, 10])
plt.show()
```
